## 2. Explore your Datasets

Now that you have your list objects created and cleaned, let's explore the descriptive statistics of these two files and make subsequent observations on our data.

```python
# TODO: Calculate the standard deviation of both the control and treatment group

std_control_data = stat.stdev(cleaned_control_data)
std_treatment_data = stat.stdev(cleaned_treatment_data)

# TODO: print your findings

print(std_control_data)
print(std_treatment_data)
```

```python
# TODO: Calculate the mean of both the control and treatment group
mean_control_data = stat.mean(cleaned_control_data)
mean_treatment_data = stat.mean(cleaned_treatment_data)

# TODO: print your findings
print(mean_control_data)
print(mean_treatment_data)
```

```python
# TODO: Are there any outliers in your control group? (use 2 * stdev to find these outliers)
lower_bound = mean_control_data - (2 * std_control_data)
upper_bound = mean_control_data + (2 * std_control_data)
outliers_control = [x for x in cleaned_control_data if x < lower_bound or x > upper_bound]


print(outliers_control)
```

```python
# TODO: Are there any outliers in your treatment group?
lower_bound = mean_treatment_data - (2 * std_treatment_data)
upper_bound = mean_treatment_data + (2 * std_treatment_data)
outliers_treatment = [x for x in cleaned_treatment_data if x < lower_bound or x > upper_bound]


print(outliers_treatment)
```
